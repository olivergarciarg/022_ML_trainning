## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="simple plot", xlabel="x-axis", ylabel="y-axis")

# 5. save & show
fig.savefig("images/sample-plot.png")
```

## making figures with numpy arrays

```python
import numpy as np
```

```python
# create data
x = np.linspace(0, 10, 101)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make plot from dictionary
prices = {"almond": 10, "peanut": 8, "cashew": 12}
fig, ax = plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title="butter prices", ylabel="price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()),list(prices.values()))
ax.set(title="butter prices", xlabel="price");
```

```python
# make some data for histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# 2 options for subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(), prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(prices.keys(), prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
# another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# hist
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6));
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target', ax=ax);
# ax.set_xlim([45,100])
```

```python
# OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))
# plot
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"]);
# customize
ax.set(title="heart disease",xlabel="age",ylabel="cholesterol");
# add legend
ax.legend(*scatter.legend_elements(), title="Target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
# plot
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"]);
# customize
ax0.set(title="heart disease",xlabel="age",ylabel="cholesterol");
# add legend
ax0.legend(*scatter.legend_elements(), title="Target");
# add a horizontal line
ax0.axhline(over_50["chol"].mean(), linestyle="--");
# plot
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"],c=over_50["target"]);
# customize
ax1.set(title="heart disease",xlabel="age",ylabel="thalach");
# add legend
ax1.legend(*scatter1.legend_elements(), title="Target");
# add a horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle="--");
# add tilte to figure
fig.suptitle("heart disease analysis", fontsize=16, fontweight="bold");
```
